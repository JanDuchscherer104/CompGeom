```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

from qhull_analyzer import QhullAnalyzerConfig
```

```python
config = QhullAnalyzerConfig(
    num_points=sorted([j * 10 ** i for j in (1, 2, 5) for i in range(1,9)])
)
experiment = config.setup()
```

```python
df = experiment.run()
df.info()
```

```python
df.head()
```

```python
sns.set_theme(style="whitegrid")
plt.figure(figsize=(12, 6))
sns.lineplot(
    data=df,
    x="num_points",
    y="runtime",
    hue="rbox_options",
    # style="qhull_options",
    markers=True,
    dashes=False,
)
# plt.yscale("log")
plt.xscale("log")
plt.title("Qhull Runtime Analysis")
plt.xlabel("Number of Points")
plt.ylabel("Runtime (seconds)")
plt.legend(title="Dimensions")
plt.show()
```
